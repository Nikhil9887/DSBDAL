```python
print("heloo")
import numpy as np
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Temp/Dsdba/Titanic-Dataset.csv')
df.head(1000)
```

```python
mean_age = df['Age'].mean()
print(mean_age)
```

```python
df['Age'].fillna(mean_age , inplace=True)
```

```python

import seaborn as sns
df.plot()
```

```python
df.plot.scatter(x = "Survived" , y = "Pclass" )
```

```python
df.plot.scatter(x = "Age" , y = "Survived" )
```

```python
import matplotlib.pyplot as plt


Pclass = df['Pclass'].dropna()


plt.figure(figsize=(8, 6))
plt.hist(Pclass , bins=100, color='skyblue', edgecolor='black')
plt.title('Histogram of Pclass')
plt.xlabel('Pclass')
plt.ylabel('Frequency')
plt.grid(axis='y')  # Show gridlines on y-axis
# plt.tight_layout()
# plt.show()
```

```python

Survived = df['Survived'].dropna()


plt.figure(figsize=(8, 6))
plt.hist(Survived , bins=100, color='skyblue', edgecolor='black')
plt.title('Histogram of Survived')
plt.xlabel('Survived')
plt.ylabel('Frequency')
plt.grid(axis='y')  # Show gridlines on y-axis
# plt.tight_layout()
# plt.show()
```

```python
Age = df['Age'].dropna()


plt.figure(figsize=(8, 6))
plt.hist(Age , bins=100, color='skyblue', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y')  # Show gridlines on y-axis
# plt.tight_layout()
# plt.show()
```
