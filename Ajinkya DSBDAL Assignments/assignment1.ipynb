```python
# mine
import numpy as np
Q1 = np.percentile(df['reading score'], 25, interpolation='midpoint')    # 1st Quartile (25 %ile)
Q3 = np.percentile(df['reading score'], 75, interpolation='midpoint')    # 3rd Quartile (75 %ile)
IQR = Q3 - Q1    
```

```python
upper = Q3 + 1.5*IQR        # upper bound
lower = Q1 - 1.5*IQR        # lower bound
```

```python
# Remove Values (Outliers) that are below the Lower Bound and above the Upper Bound

df_filtered_IQR = df[(df['reading score'] <= upper) & (df['reading score'] >= lower)]

outliers_IQR = df[(df['reading score'] >= upper) | (df['reading score'] <= lower)]

df_filtered_IQR
```
