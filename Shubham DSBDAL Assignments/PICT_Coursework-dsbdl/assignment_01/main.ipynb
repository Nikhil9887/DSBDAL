## 5. Data Formatting and Normalization


## 5.1. Handle missing values

The following can be seen as standard guideline for using mean, median or mode for replacing the missing values:

* Mean imputation is often used when the missing values are numerical and the distribution of the variable is approximately normal.
* Median imputation is preferred when the distribution is skewed, as the median is less sensitive to outliers than the mean.
* Mode imputation is suitable for categorical variables or numerical variables with a small number of unique values.

```
df.fillna(df.mean())
df.fillna(df.median())
df["salary"] = df["salary"].fillna(df["salary"].mode()[0])
```

Source: https://vitalflux.com/pandas-impute-missing-values-mean-median-mode/

### 5.1.1. Change missing values of `workclass` with mode of the feature

````python
feature_values = ds.workclass.value_counts()
mode_feature = feature_values.index.tolist()[ feature_values.argmax() ]
ds.loc[ ds.workclass == "?" , "workclass" ] = mode_feature
````

### 5.1.2. Remove records with unknown `occupation` 

````python
ds = ds.loc[ ds.occupation != "?" ]
````

### 5.1.3. Change missing values of `native-country` with mode of the feature

````python
feature_values = ds[ "native-country" ].value_counts()
mode_feature = feature_values.index.tolist()[ feature_values.argmax() ]
ds.loc[ ds[ "native-country" ] == "?" , "native-country" ] = mode_feature
````

## 5.2 Data Normalization

````python
def min_max_normalize( name ):
    global ds
    ds[name] = ds[name].astype( 'float32' )
    ds.loc[ : , name ] = (ds[ name ] - ds[ name ].min()) / ( ds[ name ].max() - ds[ name ].min() )

min_max_normalize( "age" ) 
min_max_normalize( "fnlwgt" ) 
min_max_normalize( "capital-gain" ) 
min_max_normalize( "capital-loss" ) 
min_max_normalize( "hours-per-week" ) 
````

## 6. Categorical variables to quantitative variables

### 6.1. Encoding categorical features


````python
def categorical_encoding( feature ):
    global ds
    values = ds[ feature ].unique().tolist()
    for i , val in enumerate( values ):
        ds.loc[ ds[ feature ] == val , feature ] = i
    ds[feature] = ds[feature].astype( 'int32' )

# sex and income have binary encodings    
categorical_encoding( "sex" )
categorical_encoding( "income" )

# other features have categorical
categorical_encoding( "race" )
categorical_encoding( "workclass" )
categorical_encoding( "marital-status" )
categorical_encoding( "occupation" )
categorical_encoding( "relationship" )
categorical_encoding( "native-country" )
````

````python
# education-num is the numerical encoding of `education`
# remove `education` and rename `education-num` to `education`
ds = ds.drop( [ "education" ] , axis=1 )
ds = ds.rename( { "education-num": "education" } )
````
