```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Titanic Dataset Summary:-




```python
df = pd.read_csv("./Titanic-Dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

# Correlation b/w Count of Survived and Sex

```python
# Count Plot

sns.catplot(x ="Sex", hue ="Survived", kind ="count", data = df)
plt.show()
```

# Correlation b/w Passenger Class and count of Survived

```python
# Group the dataset by Pclass and Survived and then unsack them
group = df.groupby(['Pclass', 'Survived'])
pclass_survived = group.size().unstack()

sns.heatmap(pclass_survived, annot = True, fmt ="d")
plt.show()
```

# Correlation b/w Age, Sex and Distribution of Survived

```python
# Violinplot Displays distribution of data across all levels of a category.
sns.violinplot(x ="Sex", y ="Age", hue ="Survived", data = df, split = True)
plt.show()
```

# Correlation b/w Fare and Survival Probability

```python
# Divide Fare into 4 bins
df['Fare_Range'] = pd.qcut(df['Fare'], 4)
 
# Barplot - Shows approximate values based 
# on the height of bars.
sns.barplot(x ='Fare_Range', y ='Survived', data = df)
plt.show()
```

# Correlation b/w City Embarked on, Passenger Class and Count of Survived

*Cities Embarked on are :- Cherbourg (C), Queenstown (Q), Southhampton (S)*

```python
# Countplot
sns.catplot(x ='Embarked', hue ='Survived', kind ='count', col ='Pclass', data = df)
plt.show()
```
